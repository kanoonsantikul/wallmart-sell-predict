```python
import pandas as pd
import numpy as np
```

```python
stores = pd.read_csv('stores.csv', sep=',',header='infer')
```

```python
feature = pd.read_csv('features.csv', sep=',',header='infer')

storeType = np.zeros(len(feature))
feature.insert(1, "Type", storeType)
storeType = list(stores.Type)

newfeature = pd.DataFrame()

for store_num in range(len(stores.index)):
    temp = feature[feature.Store == store_num + 1] 
    temp.Type = storeType[store_num]
    newfeature = newfeature.append(temp)

feature = newfeature
feature
```

```python
train = pd.read_csv('train.csv', sep=',',header='infer')
train['Type'] = 'A'
train['Temperature'] = 0.0
train['Fuel_Price'] = 0.0
train['MarkDown1'] = 0.0
train['MarkDown2'] = 0.0
train['MarkDown3'] = 0.0
train['MarkDown4'] = 0.0
train['MarkDown5'] = 0.0
train['CPI'] = 0.0
train['Unemployment'] = 0.0

out = pd.DataFrame()
for i in range(150):
    print(i)
    day = feature.iloc[i]
    train.loc[(train.Date == day.Date) & (train.Store == day.Store), 'Type'] = day.Type
    train.loc[(train.Date == day.Date) & (train.Store == day.Store), 'Temperature']  = day.Temperature
    train.loc[(train.Date == day.Date) & (train.Store == day.Store), 'Fuel_Price']  = day.Fuel_Price
    train.loc[(train.Date == day.Date) & (train.Store == day.Store), 'MarkDown1']  = day.MarkDown1
    train.loc[(train.Date == day.Date) & (train.Store == day.Store), 'MarkDown2']  = day.MarkDown2
    train.loc[(train.Date == day.Date) & (train.Store == day.Store), 'MarkDown3']  = day.MarkDown3
    train.loc[(train.Date == day.Date) & (train.Store == day.Store), 'MarkDown4']  = day.MarkDown4
    train.loc[(train.Date == day.Date) & (train.Store == day.Store), 'MarkDown5']  = day.MarkDown5
    train.loc[(train.Date == day.Date) & (train.Store == day.Store), 'CPI'] = day.CPI
    train.loc[(train.Date == day.Date) & (train.Store == day.Store), 'Unemployment'] = day.Unemployment
    
train
```
