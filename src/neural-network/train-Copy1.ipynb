```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
import os


# model_name = input('model name: ')
all_data = pd.read_csv('../../train_merged.csv')

all_data = all_data[['Store', 'Dept', 'Date', 'Weekly_Sales']]
all_data['WeekNum'] = pd.to_datetime(all_data['Date']).dt.week
del all_data['Date']
```

```python
test = pd.read_csv('../../data/sampleSubmission.csv', sep='_')
test['WeekNum'] = pd.to_datetime(test['Date']).dt.week
del test['Date']
```

```python
# all_data = pd.get_dummies(all_data)

all_data['Store'] = all_data['Store'].astype('O')
all_data['Dept'] = all_data['Dept'].astype('O')
all_data['WeekNum'] = all_data['WeekNum'].astype('O')
all_data = pd.get_dummies(all_data)
```

```python
test['Store'] = test['Store'].astype('O')
test['Dept'] = test['Dept'].astype('O')
test['WeekNum'] = test['WeekNum'].astype('O')
test = pd.get_dummies(test)
```

```python
for i in range(1, 53):
    col = 'WeekNum_'+str(i)
    if col not in list(test.columns):
        test[col] = 0
```

```python
Y = all_data['Weekly_Sales'].values
del all_data['Weekly_Sales']

index = list(all_data.columns)
index
all_data = all_data.reindex(columns=index)
test = test.reindex(columns=index)
```

```python
from keras.models import Sequential
from keras.layers import Dense
from keras.optimizers import Adam
import h5py


def build_model(shape):
    model = Sequential()
    model.add(Dense(64, input_shape=shape, activation='relu'))
    model.add(Dense(64, activation='relu'))
    model.add(Dense(64, activation='relu'))
    model.add(Dense(1))
    model.compile(optimizer=Adam(lr=0.01), loss='mean_absolute_error')
    return model

def evaluate(model, x, y):
    predicted = model.predict(x)[:, 0]
    error = np.absolute(predicted - y)
    mae = sum(error) / len(error)
    return {'predicted':predicted, 'mae':mae}
```

```python
X = all_data.values

EPOCHS = int(input('Epochs :'))

model = build_model((X.shape[1],))
history = model.fit(X, Y,
            epochs=EPOCHS,
            batch_size=X.shape[0])
```

```python
result = model.predict(test.values)[:, 0]
result
```

```python
submit = pd.read_csv('../../data/sampleSubmission.csv')
submit['Weekly_Sales'] = result
submit.columns = ['Id', 'Weekly_Sales']
submit.to_csv('../../data/nn_weeknum_submit03.csv', index=False, na_rep='0')
```
